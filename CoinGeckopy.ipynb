```python
import pandas as pd
from pycoingecko import CoinGeckoAPI

# Initialize CoinGecko API client
cg = CoinGeckoAPI()

# Fetch market data for multiple coins (example: Bitcoin, Ethereum)
coin_market_data = cg.get_coins_markets(vs_currency="usd")

# Convert to DataFrame
df = pd.DataFrame(coin_market_data)

# Print available columns (to verify column names before dropping)
print("Available columns:", df.columns.tolist())

# Drop unnecessary columns (only if they exist)
columns_to_drop = [
    "id", "symbol", "image", "high_24h", "low_24h", "price_change_24h",
    "price_change_percentage_24h", "market_cap_change_24h",
    "market_cap_change_percentage_24h", "fully_diluted_valuation",
    "ath_date", "ath_change_percentage", "atl_change_percentage",
    "atl_date", "roi"
]

# Drop columns that exist in the DataFrame

df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

# Save to CSV
df.to_csv("crypto_market_data.csv", index=False)

print("CSV file 'crypto_market_data.csv' saved successfully!")
```

```python
import pandas as pd
from pycoingecko import CoinGeckoAPI

# Initialize CoinGecko API client
cg = CoinGeckoAPI()

# Set the number of coins per request (maximum is 250)
per_page = 250
total_coins = 4000
coins_data = []

# Fetch multiple pages to get a total of 4000 coins
for page in range(1, (total_coins // per_page) + 2):  # Add 2 to include the last page (if needed)
    print(f"Fetching page {page} of {total_coins // per_page + 1}...")
    coin_market_data = cg.get_coins_markets(vs_currency="usd", per_page=per_page, page=page)
    coins_data.extend(coin_market_data)

# Convert to DataFrame
df = pd.DataFrame(coins_data)

# Drop unnecessary columns (only if they exist)
columns_to_drop = [
    "id", "symbol", "image", "high_24h", "low_24h", "price_change_24h",
    "price_change_percentage_24h", "market_cap_change_24h",
    "market_cap_change_percentage_24h", "fully_diluted_valuation",
    "ath_date", "ath_change_percentage", "atl_change_percentage",
    "atl_date", "roi"
]

df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

# Save to CSV
df.to_csv("crypto_market_data_4000.csv", index=False)

print("CSV file 'crypto_market_data_4000.csv' saved successfully!")
```
